### Data Preprocessing

From the iris dataset we need to divide the data into target variable and independent variable.Before that we need to convert the spiece into numeric format.

```python
iris.drop(['Id'], axis = 'columns',inplace = True)
iris.head()
```

```python
# Convert text values into numeric
label_enco = LabelEncoder()
```

```python
# reason ml model works only with numbers
#since species are in text we need to convert it 

iris['Species'] = label_enco.fit_transform(iris['Species'])
```

```python
# we are assigning "x" as the independent variables and "y" as the target variable
x = iris.drop(['Species'],axis = 'columns')
y = iris['Species']
```

```python
y
```

```python
x
```

### ML model

For making the ML model , we already imported the library "tree"

```python
#creating  the model 
model = tree.DecisionTreeClassifier()
```

```python
#training the model

print('Decision Tree Classifer Created')

model.fit(x,y)
```

### Visualization of the Model

Now we are going to display how the tree look in raw

```python
tree.plot_tree(model)
```

```python
#let's color it up

feature_cols = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
class_names = ['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = feature_cols, 
               class_names=class_names,
               filled = True);
fig.savefig('Deicision_tree_iris.png')
```

```python
model.score(x,y)
```
